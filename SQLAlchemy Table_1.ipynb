```python
import sqlalchemy
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import sessionmaker, scoped_session
```

```python
import pandas as pd
import os
```

```python
os.path.join(os.getcwd(), 'Table_1.db.sqlite3')
```

```python
"""Create a database file."""
database_file = os.path.join(os.path.abspath(os.getcwd()), 'Table_1.db.sqlite3')
```

```python
database_file
```

```python
"""Creat an engine object."""
engine = sqlalchemy.create_engine('sqlite:///' + database_file, echo = True)
```

```python
"""Creat a session."""
db_session = scoped_session(
    sessionmaker(
        autocommit = False,
        autoflush = False,
        bind = engine
    )
)
```

```python
"""Create a base class."""
Base = declarative_base()
Base.query = db_session.query_property()
```

```python
"""Create a model class."""
class Table(Base):
    __tablename__ = 'Table_1'
    id = Column(Integer, primary_key = True)
    x = Column(Float, unique = False)
    y1 = Column(Float, unique = False)
    y2 = Column(Float, unique = False)
    y3 = Column(Float, unique = False)
    y4 = Column(Float, unique = False)    
    
    def __init(self, x = None, y1 = None, y2 = None, y3 = None, y4 = None):
        self.x = x
        self.y1 = y1
        self.y2 = y2
        self.y3 = y3
        self.y4 = y4
```

```python
"""Creat a database."""
Base.metadata.create_all(bind = engine)
```
