```python
class B:
    def __init__(self, a):
        print('B init')
        self.a = a

    def some_method(self):
        print(f'a: {self.a}')

class D(B):
    def __init__(self, a, b):
        print('D init')
        super().__init__(a)
        self.b = b

    def another_method(self):
        self.some_method()
        print(f'b: {self.b}')
```

```python
import unittest
from unittest import TestCase

class TestTokuten_data(unittest.TestCase):
    
    def setUp(self):
        print ("setup")
        self.eq = Tokuten_data()
        
    def test_heikin_cal(self):
        """test method of heikin_cal"""
        expected = (1.5)
        actual = self.eq.heikin_cal()
        
        self.assertEqual(expected, actual)
        
    def tearDown(self):
        print("tearDown")
        del self.eq
```

```python
def tashizan(a, b):
    """calculate a+b

    :param a: value 1(int)
    :param b: value 2(inte)
    :return: a + b
    """
    return a + b


def hikizan(a, b):
    """calculate a - b

    :param a:  value 1(int)
    :param b:  value 2(int)
    :return:  a - b
    """
    return a - b
```

```python
import unittest
from unittest import TestCase

class TestKeisan(unittest.TestCase):
    """test class of keisan.py
    """
    def setUp(self):
        print("setup")

    def test_tashizan(self):
        """test method for tashizan
        """
        value1 = 2
        value2 = 6
        expected = 8
        actual = tashizan(value1, value2)
        self.assertEqual(expected, actual)
        
    def tearDown(self):
        print("tearDown")


class TestKeisan2(unittest.TestCase):
    """test class of keisan2.py"""
    def setUp(self):
        print("setup")
        
    def test_hikizan(self):
        """test method for hikizan
        """
        value1 = 2
        value2 = 12
        expected = -10
        actual = hikizan(value1, value2)
        self.assertEqual(expected, actual)
        
    def tearDown(self):
        print("tearDown")
```
