```python
import sqlalchemy
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.orm import sessionmaker, scoped_session
```

```python
import pandas as pd
import os
```

```python
os.path.join(os.getcwd(), 'Table_3.db.sqlite3')
```

```python
"""Create a database file."""
database_file = os.path.join(os.path.abspath(os.getcwd()), 'Table_3.db.sqlite3')
```

```python
database_file
```

```python
"""Create an engine obeject."""
engine = sqlalchemy.create_engine('sqlite:///' + database_file, echo = True)
```

```python
"""Create a session."""
db_session = scoped_session(
    sessionmaker(
        autocommit = False,
        autoflush = False,
        bind = engine
    )
)
```

```python
"""Create a base class."""
Base = declarative_base()
Base.query = db_session.query_property()
```

```python
"""Create a model class."""
class Table(Base):
    __tablename__ = 'Table_3'
    id = Column(Integer, primary_key = True)
    x = Column(Float, unique = False)
    y = Column(Float, unique = False)
    Delta_Y_test_func = Column(Float, unique = False)
    No_of_ideal_func = Column(String, unique = False)
    
    def __init(self, x = None, y = None, Delta_Y_test_func = None, No_of_ideal_func = None):
        self.x = x
        self.y = y
        self.Delta_Y_test_func = Delta_Y_test_func
        self.No_of_ideal_func = No_of_ideal_func
```

```python
"""Create a database."""
Base.metadata.create_all(bind = engine)
```
