```python
"""This is a class to combine the three functions showned above to find the four functions."""
class Ideal_Four:
    
    def __init__(self):
        self.ideal = ideal
        
    def MSE(self, train):
        mse = []
        for i in range(1, 51):
            i = str(i)
            mse.append(sum((train - self.ideal["y" + i]) ** 2) /\
                       len(train))

        return mse.index(min(mse)) + 1, min(mse)
    
    def MSE_ALL(self, train):
        mse_all = []
        for i in range(1, 5):
            i = str(i)
            mse_all.append(self.MSE(train["y" + i]))

        return mse_all
    
    def Ideal_Dataset(self, train):
        dataset = []
        for i in range(4):
            dataset.append(self.ideal\
                           ["y" + str(self.MSE_ALL(train)[i][0])])

        return dataset
```

```python
"""Create a class to produce Delta and No by combination of the two functions."""
class Delta_No:
    
    def __init__(self, Ideal_four, test):
        self.Ideal_four = Ideal_four
        self.test = test
        
    def TEST(self, train, j):
        Test = []
        for i in range(4):
            dev = self.Ideal_four[i][self.Ideal_four[i].index == self.test.iloc[j, 0]].iloc[0] \
    - self.test.iloc[j, 1]
            Test.append(dev)

        Test_min = min(Test, key = abs)
        Test_index = Test.index(Test_min)

        return Test_min, Test_index
    
    def TEST_ALL(self, train):
        Test_all_min = []
        Test_all_index = []
        for i in range(100):
            Test_min = TEST(train, i)[0]
            Test_index = TEST(train, i)[1]
            Test_all_min.append(Test_min)
            Test_all_index.append(Test_index)
        return Test_all_min, Test_all_index
```
