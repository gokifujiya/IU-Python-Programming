```python
train = pd.read_csv('train.csv', index_col = 0)
train.head()
```

```python
ideal = pd.read_csv('ideal.csv', index_col = 0)
ideal.head()
```

```python
def MSE(train):
    mse = []
    for i in range(1, 51):
        i = str(i)
        mse.append(sum((train - ideal["y" + i]) ** 2) / len(train))
        
    return mse.index(min(mse)) + 1, min(mse)
```

```python
def MSE_ALL(train):
    mse_all = []
    for i in range(1, 5):
        i = str(i)
        mse_all.append(MSE(train["y" + i]))
    
    return mse_all
```

```python
def Ideal_Dataset(train):
    dataset = []
    for i in range(4):
        dataset.append(ideal["y" + str(MSE_ALL(train)[i][0])])
        
    return dataset
```
